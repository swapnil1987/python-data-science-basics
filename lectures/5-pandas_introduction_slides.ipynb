## Series: Single Column of Data

A Series is:
- A single column of data
- With row labels (called the *index*) for each observation



## Creating a Series

US unemployment rate every other year starting in 1995

```python
values = [5.6, 5.3, 4.3, 4.2, 5.8, 5.3, 4.6, 7.8, 9.1, 8., 5.7]
years = list(range(1995, 2017, 2))

unemp = pd.Series(data=values, index=years, name="Unemployment")
unemp
```

## Series Components

Access the index and values separately

```python
unemp.index
```

```python
unemp.values
```

## Viewing Data: `.head()` and `.tail()`

```python
unemp.head()
```

```python
unemp.tail()
```

## Basic Plotting

```python
unemp.plot()
```

## Unique Values

```python
unemp.unique()
```

## Indexing with `.loc`

Select particular elements using `.loc[index_items]`

```python
# Single element
unemp.loc[1995]
```

```python
# Multiple elements
unemp.loc[[1995, 2005, 2015]]
```

## DataFrame: Multiple Columns

A DataFrame is:
- Multiple Series stacked side by side as columns
- Similar to an Excel sheet or SQL table
- Has both row labels (index) and column labels



## Creating a DataFrame

Unemployment rate by US region

```python
data = {
    "NorthEast": [5.9,  5.6,  4.4,  3.8,  5.8,  4.9,  4.3,  7.1,  8.3,  7.9,  5.7],
    "MidWest": [4.5,  4.3,  3.6,  4. ,  5.7,  5.7,  4.9,  8.1,  8.7,  7.4,  5.1],
    "South": [5.3,  5.2,  4.2,  4. ,  5.7,  5.2,  4.3,  7.6,  9.1,  7.4,  5.5],
    "West": [6.6, 6., 5.2, 4.6, 6.5, 5.5, 4.5, 8.6, 10.7, 8.5, 6.1],
    "National": [5.6, 5.3, 4.3, 4.2, 5.8, 5.3, 4.6, 7.8, 9.1, 8., 5.7]
}

unemp_region = pd.DataFrame(data, index=years)
unemp_region
```

## DataFrame Components

```python
unemp_region.index
```

```python
unemp_region.values
```

## DataFrame: `.head()` and `.tail()`

```python
unemp_region.head()
```

## Plotting DataFrames

Each column gets its own line

```python
unemp_region.plot()
```

## DataFrame Indexing with `.loc`

Select subsets of rows AND columns

```python
# Single element
unemp_region.loc[1995, "NorthEast"]
```

```python
# Multiple rows, single column
unemp_region.loc[[1995, 2005], "South"]
```

## More Indexing Examples

```python
# Single row, multiple columns
unemp_region.loc[1995, ["NorthEast", "National"]]
```

```python
# All rows, single column
unemp_region.loc[:, "NorthEast"]
```

```python
# Shorthand: extract whole column without .loc
unemp_region["MidWest"]
```

## Computations with Columns

pandas applies operations to all rows at once

```python
# Divide by 100 to convert from percent to rate
unemp_region["West"] / 100
```

```python
# Find maximum
unemp_region["West"].max()
```

## Column Arithmetic

```python
# Find difference between two columns
unemp_region["West"] - unemp_region["MidWest"]
```

## Correlation Analysis

```python
# Correlation between two columns
unemp_region.West.corr(unemp_region["MidWest"])
```

```python
# Correlation matrix for all column pairs
unemp_region.corr()
```

## Data Types (dtypes)

pandas distinguishes between:
- Booleans (`bool`)
- Floating point numbers (`float64`)
- Integers (`int64`)
- Dates (`datetime`)
- Categorical data (`categorical`)
- Everything else, including strings (`object`)

```python
unemp.dtype
```

```python
unemp_region.dtypes
```

## Why dtypes Matter

Incorrect dtypes can cause unexpected behavior

```python
# Convert South column to string
str_unemp = unemp_region.copy()
str_unemp["South"] = str_unemp["South"].astype(str)
str_unemp.dtypes
```

## dtype Problems Example

```python
str_unemp.head()
```

```python
# Unexpected result: strings concatenate instead of adding!
str_unemp.sum()
```

## Creating New Columns

Syntax: `df["New Column Name"] = new_values`

```python
# Unweighted mean across regions
unemp_region["UnweightedMean"] = (unemp_region["NorthEast"] +
                                  unemp_region["MidWest"] +
                                  unemp_region["South"] +
                                  unemp_region["West"])/4

unemp_region.head()
```

## Changing Values

Use sparingly: `df.loc[index, column] = value`

```python
unemp_region.loc[1995, "UnweightedMean"] = 0.0
unemp_region.head()
```

## Renaming Columns

Use a dictionary with old names as keys, new names as values

```python
names = {"NorthEast": "NE",
         "MidWest": "MW",
         "South": "S",
         "West": "W"}

unemp_region.rename(columns=names)
```

## Making Changes Permanent

Two options:
1. Assign back: `df = df.rename(columns=rename_dict)`
2. Use `inplace=True`: `df.rename(columns=rename_dict, inplace=True)`

**Recommendation:** Use option 1 until comfortable with pandas

```python
unemp_shortname = unemp_region.rename(columns=names)
unemp_shortname.head()
```
